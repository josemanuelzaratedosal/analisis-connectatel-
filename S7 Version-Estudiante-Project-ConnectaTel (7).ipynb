**🎯 Objetivo:**

### 3.2 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Decidir qué hacer con los valores nulos según su proporción y relevancia.

**Instrucciones:**
- Verifica si los nulos en `duration` y `length` son **MAR**(Missing At Random) revisando si dependen de la columna `type`.  
  Si confirmas que son MAR, **déjalos como nulos** y justifica la decisión.

```python
# 1. Verificación de nulos para duration y length agrupados por type
print("Distribución de nulos según el tipo de servicio:")
print(usage.groupby('type')[['duration', 'length']].apply(lambda x: x.isna().sum()))

# 2. Verificar si hay mensajes ('text') con duración o llamadas ('call') sin duración
mensajes_con_duracion = usage[(usage['type'] == 'text') & (usage['duration'] > 0)]
llamadas_sin_duracion = usage[(usage['type'] == 'call') & (usage['duration'].isna())]

print(f"\nMensajes con duración registrada: {len(mensajes_con_duracion)}")
print(f"Llamadas con duración nula: {len(llamadas_sin_duracion)}")
```

```python
# Verificación MAR en usage para length
print("Distribución de nulos en length según el tipo de servicio:")
print(usage.groupby('type')['length'].apply(lambda x: x.isna().sum()))

# La prueba real de MAR: Las llamadas DEBEN tener length nulo y los mensajes NO.
# Verificamos si hay llamadas que SÍ tengan algo en length (esto sería el error)
llamadas_con_length = usage[(usage['type'] == 'call') & (usage['length'].notna())]
print(f"\nLlamadas con longitud de texto registrada (debería ser 0): {len(llamadas_con_length)}")

# Verificamos si hay mensajes que NO tengan length (esto también sería un error)
mensajes_sin_length = usage[(usage['type'] == 'text') & (usage['length'].isna())]
print(f"Mensajes sin longitud registrada: {len(mensajes_sin_length)}")
```

Diagnóstico de nulos en duration y length
Análisis de duration: Los valores nulos aparecen exclusivamente cuando el tipo de servicio es 'text'. Esto es lógicamente correcto, ya que un mensaje de texto no tiene una duración en minutos.

Análisis de length: Los valores nulos aparecen únicamente cuando el tipo de servicio es 'call'. Esto también es correcto, pues una llamada de voz no posee una longitud de caracteres.

## 🧩Paso 4: Summary statistics de uso por usuario


### 4.1 Agrupación por comportamiento de uso

🎯**Objetivo**: Resumir las variables clave de la tabla `usage` **por usuario**, creando métricas que representen su comportamiento real de uso histórico. 

**Instrucciones:**: 
1. Construye una tabla agregada de `usage` por `user_id` que incluya:
- número total de mensajes  
- número total de llamadas  
- total de minutos de llamadas

2. Renombra las columnas para que tengan nombres claros:  
- `cant_mensajes`  
- `cant_llamadas`  
- `cant_minutos_llamada`
3. Combina esta tabla con `users`.

```python
usage['is_text'] = 1
usage['is_call'] = 1

usage_agg = usage.groupby('user_id').agg({
    'is_text': 'sum',
    'is_call': 'sum',
    'duration': 'sum'
}).reset_index()

print(usage_agg.head(3))
```

### 5.2 Identificación de Outliers

🎯 **Objetivo:**  
Detectar valores extremos en las variables clave de **uso** y **clientes** que podrían afectar el análisis, y decidir si requieren limpieza o revisión adicional.

**Instrucciones:**  
- Usa **boxplots** para identificar visualmente outliers en las siguientes columnas:  
  - `age` 
  - `cant_mensajes`
  - `cant_llamadas`
  - `total_minutos_llamada`  
- Crea un **for** para generar los 4 boxplots automáticamente.
<br>

- Después de crear los gráfico, responde si **existen o no outliers** en las variables.  
- Si hay outliers, crea otro bucle para calcular los límites de esas columnas usando el **método IQR** y decide qué hacer con ellos.
  - Si solamente hay outliers de un solo lado, no es necesario calcular ambos límites.

**Hint:**
- Dentro del bucle, usa `plt.title(f'Boxplot: {col}')` para que el título cambie acorde a la columna.

```python
# 1. Boxplots automáticos
columnas_numericas = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

for col in columnas_numericas:
    plt.figure(figsize=(10, 4))
    sns.boxplot(data=user_profile, x=col, hue='plan', palette=['skyblue', 'green'])
    plt.title(f'Boxplot: {col} por Tipo de Plan')
    plt.xlabel(col)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

# 2. Respuesta a la pregunta:
# "Existen outliers principalmente en 'cant_mensajes' y 'cant_minutos_llamada' 
# (hacia la derecha). En 'age' no se observan valores atípicos significativos."

# 3. Cálculo de límites con Método IQR (Paso obligatorio según tus instrucciones)
cols_con_outliers = ['cant_mensajes', 'cant_minutos_llamada']

for col in cols_con_outliers:
    Q1 = user_profile[col].quantile(0.25)
    Q3 = user_profile[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5 * IQR
    
    print(f"Límite superior para {col}: {limite_superior}")
    print(f"Registros por encima del límite: {len(user_profile[user_profile[col] > limite_superior])}\n")
```

💡Insights: 
- Age: No presenta outliers. La distribución es uniforme y todos los datos caen dentro de los bigotes.
- cant_mensajes: Presenta outliers hacia la derecha. Hay usuarios con un consumo de mensajes muy superior al promedio.
- cant_llamadas: No suele presentar outliers significativos (o son mínimos), ya que la distribución es más simétrica.
- cant_minutos_llamada: Presenta algunos outliers hacia la derecha, representando usuarios con llamadas muy largas.

```python
# Calcular límites con el método IQR
columnas_limites = ['cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

for col in columnas_limites:
    Q1 = user_profile[col].quantile(0.25)
    Q3 = user_profile[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculamos solo el límite superior ya que el sesgo es a la derecha
    limite_superior = Q3 + (1.5 * IQR)
    
    # Contar cuántos registros superan ese límite
    conteo_outliers = (user_profile[col] > limite_superior).sum()
    
    print(f"Variable: {col}")
    print(f"  > Límite superior: {limite_superior:.2f}")
    print(f"  > Cantidad de outliers: {conteo_outliers}")
    print("-" * 30)
```

### 6.1 Segmentación de Clientes Por Uso

🎯 **Objetivo:** Clasificar a cada usuario en un grupo de uso (Bajo uso, Uso medio, Alto uso) basándose en la cantidad de llamadas y mensajes registrados.

**Instrucciones:**  
- Crea una nueva columna llamada `grupo_uso` en el dataframe `user_profile`.
- Usa comparaciones lógicas (<, >) para evaluar las condiciones de llamadas y mensajes y asigna:
  - `'Bajo uso'` cuando llamadas < 5 y mensajes < 5
  - `'Uso medio'` cuando llamadas < 10 y mensajes < 10
  - `'Alto uso'` para el resto de casos

```python
import numpy as np

# 1. Definir las condiciones lógicas
condiciones = [
    (user_profile['cant_llamadas'] < 5) & (user_profile['cant_mensajes'] < 5), # Bajo uso
    (user_profile['cant_llamadas'] < 10) & (user_profile['cant_mensajes'] < 10) # Uso medio
]

# 2. Definir los nombres de los grupos
opciones = ['Bajo uso', 'Uso medio']

# 3. Crear la columna con np.select (el default es 'Alto uso')
user_profile['grupo_uso'] = np.select(condiciones, opciones, default='Alto uso')

# 4. Verificar distribución de los grupos
print(user_profile['grupo_uso'].value_counts())

# Mostrar resultado inicial
user_profile[['user_id', 'cant_llamadas', 'cant_mensajes', 'grupo_uso']].head()
```
